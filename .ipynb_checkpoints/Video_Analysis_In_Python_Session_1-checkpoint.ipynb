## What about getting time data?   The "fps" metadata is important.

Before we try to track the motion of your object, we need to know how to calculate the time of each successive frame of the video!  The video file contains **metadata** including the *frames per second* (fps) at which the video was taken.  Let's look at this now.

* It's important to note that **not every frame** contains a time stamp, only certain frames do.  So we can search through the video until we find the frames that have time stamps, and get the fps from these.


```python
fps_values = np.array([])

for num, image in enumerate(vid.iter_data()):  #This is a way of scrolling through the frames in the video.
        if num % int(vid._meta['fps']):
            continue   #If there's no time stamp, we don't examine the video frame any further. 
            #Go to the next one.
        else:
            fps = vid.get_meta_data()['fps']
            fps_values = np.append(fps_values, fps)

print("Mean fps: " + str( fps_values.mean()))
```

## First steps in video analysis

a. First, figure out at what frame your pendulum begins swinging such that it *is entirely within the video frame and doesn't leave the boundaries of the video*.  If you begin tracking while the pendulum is swinging out of the picture, you'll lose track of it and your data set will be hard to deal with.


b. Next, it's time to take advantage of the fact that you've marked your pendulum in red.  What we're going to do now is examine the **blue** color channel.   Select a frame of your video where the pendulum is clearly visible and display only the green color channel using  the <code>cm.gray</code> color map.
```python
# Pendulum is entirely in video starting at frame 1620
pendulum = vid.get_data(1680)
plot.imshow(pendulum[:,:,2], cmap=cm.gray)   
plot.show()
```

```python
blueFrame = pendulum[:,:,2]
redFrame = pendulum[:,:,0]
```

* **QUESTION 2**.  What color does the red pendulum mass appear to be in the blue color channel?  Why?


The red pendulum appears black in the blue color channel because a computer monitor will not show any blue light when displaying red.



### We can use this to our advantage.

**Here is the process.  we will do this to one example frame below.**

*  We convert the red and blue color channels to floating point values.  

* The blue color channel shows us the background level (which, because white, should be the same on every color channel) with a "hole" of zero-intensity values where the red pendulum is located.  We create a new array of differences between red and blue:


<code>   diffs = red_channel - green_channel </code>


The diffs array should now show only the red pendulum remaining!  


**QUESTION 3.**  Explain why / how this works?

This works because we are light correcting for the background. Since the background is more or less constant between red and blue (since it is white), it will subtract out leaving behind only the pendulum


*  On the offchance that the background still not fully subtracted out, we loop through each pixel and **set it to zero if its value is below a threshold**.  This ensures that pixels below a certain threshold of intensity will not be included in our calculation.  The trick here is to find a reasonable threshold limit.  Should that be 5.0?  20.0?  80.0?  150.0?  It will be up to you to find out.


*  We next use the "center of mass" calculation that you examined in the pre-lab to find the center of the red pendulum.  *This only works because we have subtracted the background out of our video!*


*  We finally store this location in an array.  We load up the next video frame, and do the same steps again!


---------

### We're going to begin by processing just one frame, the example frame you have picked.  Once our method works for one frame, it will (hopefully) work for every frame. 


a. First, to do meaningful calculations with this video we must convert the data in each frame into *floating point* data. Convert your frame of blue channel data to a floating point NumPy array basically with this method:


<code> my_new_array = my_array.astype(float)</code>


In the cells below, create floating-point format arrays for both the red and green channels of the example frame you've chosen:


```python
newBlue = blueFrame.astype(float)
```

```python
newRed = redFrame.astype(float)
```


b. Now, since you used a **white background**, and since each color channel in your video stores intensity values from 0 to 255,  it's safe to assume that the background in your green color channel has a value close to the value in your red color channel.   Taking the difference should subtract out the background of your video frame, leaving only the red pendulum.  Create this difference array in the cell below:


```python
diffs = newRed - newBlue
```

c. Now, just in case the background is not fully gone, we apply **a threshold test** to set array values below some threshold to zero.  You can use the "<code>ndenumerate()</code>" function which will create a loop that gives you the *value* of a pixel and the *location (index)* of the pixel:


<code>  threshold = 50  #Example threshold
        for index, value in np.ndenumerate(diff):
            if value < threshold:           #Use an IF statement.
                myarray[index] = 0.0
</code>


Now process your array in the cell below with a threshold you have chosen.  The threshold should be nearer to zero, than to 255.0  (because you want to mask out any part of the background that didn't end up quite at zero after subtracting it out!).  

```python
threshold = 60
for index, value in np.ndenumerate(diffs):
    if value < threshold:           #Use an IF statement.
        newRed[index] = 0.0
```

e.  Now, imshow() your processed array using the cm.gray color map!  *If you see a significant number of stray pixels in the background you should tweak your threshold and keep viewing the result until you find a workable value.*

```python
plot.imshow(newRed, cm.gray)
plot.show()
```

f.  **Now do the center of mass calculation on your frame and plot a blue dot (using the plot.plot() command) at the position you calculate the pendulum to be!**  Refer to your pre-lab if you need do.

```python
floatingpoint_version = newRed[:,:]/1.0

#Now we can sum up the intensities and have enough numerical resolution to do this.
summed_intensity = floatingpoint_version[:,:].sum()

R = np.array([0,0])

#Now that we've got that calculated, loop through each pixel:
for j in range(0,newRed.shape[0]):  #LOOP THROUGH Y PIXEL POSITIONS
    for i in range(0,newRed.shape[1]):  #LOOP THROUGH X PIXEL POSITIONS
        R = R + np.array([j,i])*newRed[j,i]  #MULTIPLY EACH POSITION BY INTENSITY IT CONTAINS

#Now finally get the mean by dividing by total summed intensity of ALL pixels in the rectangle:
R = R/summed_intensity

print(R)  #(Y, X) position of the centroid
```

## In the remainder of this notebook, create a for loop that will loop through your video frame by frame.  

* The **intro/tutorial video** for this lab may be a useful resource as well.

* All you need to do is perform this same calculation for each successive frame in your video, storing the calculated center of mass in a NumPy array.  At the end,  you should have a data set containing the positions of your pendulum!


* You should likewise create a time array that stores the time of each frame, using the fps value that we determined from your video.


* **Speedup hint (very important):**  If your pendulum only occupies a certain range of x or y  pixels in your video, you can do  your center of mass calculation on *just that rectangle of pixels* !


* **Speedup hint 2:**  Only loop through the video frames where something is actually happening.  If you have 15 seconds of non-useful video before and after your experiment, you'll waste precious time "processing" this otherwise.  Constrain your calculation to just the video frames where your pendulum will actually give you useful data!

```python
yposition = np.array([])
xposition = np.array([])
time = np.array([])
threshold = 100
for frameNumber in range(1620,4050):
    pendulums = vid.get_data(frameNumber)
    blueF = pendulums[500:700,:,2]
    redF = pendulums[500:700,:,0]
    newB = blueF.astype(float)
    newR = redF.astype(float)
    diff = newR - newB
    for index, value in np.ndenumerate(diff):
        if value < threshold:
            newR[index] = 0.0
    fp_vers = newR[:,:]/1.0
    sum_int = fp_vers[:,:].sum()
    Rr = np.array([0,0])
    for j in range(0,newR.shape[0]):
        for i in range(0,newR.shape[1]):
            Rr = Rr + np.array([j,i])*newR[j,i]
    Rr = Rr/sum_int
    yposition = np.append(yposition, Rr[0])
    xposition = np.append(xposition, Rr[1])
    time = np.append(time, (frameNumber/30.0)-54)
```

```python
from scipy.integrate import odeint
```

```python
def damped(stateVec, t):
    x,v = stateVec
    f = [v, -(b*v/m) - (k*x/m)]
    return f
```

```python
#Underdamped
m = 2.718
k = 1.618
b = 0.314*(4*m*k)**0.5
state = [0.0, .5]

end = 25.0
points = 5000

t = [end*float(i)/(points - 1) for i in range(points)]

solutionU = odeint(damped, state, t)
posU = solutionU[:,0]
velU = solutionU[:,1]
```

```python
plot.figure(dpi = 100)
plot.plot(t,posU,'b-', label = 'underdamped position (b = 0.314(4mk)^0.5)')
plot.plot(t,velU,'b:', label = 'underdamped velocity')
plot.xlabel("Time (s)")
plot.ylabel("Position (m) and Velocity (m/s)")
plot.suptitle("x(t) and v(t) for damped SHO", fontsize = 15)
plot.title("m = 2.718, k = 1.618, and initial state = [0.0, 0.5]")
plot.legend()
plot.show()
```

```python
def driven(stateVec, t):
    x,v = stateVec
    f = [v, F_0 * np.sin(freq * t) - (b*v/m) - (k*x/m)]
    return f
```

```python
#Driven
m = 2.718
k = 1.618
b = 0.25
F_0 = 0.1
freq = (k/m)**0.5
state = [0.0, 0.0]

end = 480.0
points = 96000

t = [end*float(i)/(points - 1) for i in range(points)]

solutionD = odeint(driven, state, t)
posD = solutionD[:,0]
velD = solutionD[:,1]

##Note: guessed sqrt(k/m) as resonant frequency
```

```python
plot.figure(dpi = 100)
plot.plot(t,posD,'b-', label = 'driven position (b = 0.25)')
plot.plot(t,velD,'b:', label = 'driven velocity')
plot.xlabel("Time (s)")
plot.ylabel("Position (m) and Velocity (m/s)")
plot.suptitle("x(t) and v(t) for damped driven SHO", fontsize = 15)
plot.title("m = 2.718, k = 1.618, freq = sqrt(k/m), and initial state = [0.0, 0.0]",
           fontsize = 10)
plot.legend()
plot.show()
```

```python
maxPos = np.array([])
maxVel = np.array([])
for i in range(0,4000,1):
    #Driven
    m = 2.718
    k = 1.618
    b = 0.25
    F_0 = 0.1
    freq = i/2000
    state = [0.0, 0.0]

    end = 480.0
    points = 96000

    t = [end*float(j)/(points - 1) for j in range(points)]

    solutionD = odeint(driven, state, t)
    posD = solutionD[:,0]
    velD = solutionD[:,1]
    maxPos = np.append(maxPos, np.max(posD[100:]))
    maxVel = np.append(maxVel, np.max(velD[100:]))
```

```python
freqRange = np.linspace(0,2,4000)
plot.plot(freqRange, maxPos)
plot.plot(freqRange, maxVel)
plot.show()
```

```python
def dpend(stateVec, t):
    th,w = stateVec
    tau = [w, (-b*w/m)-(k*np.sin(th))]
    return tau
```

```python
#Driven
m = 0.1
k = 4.3 #Allegedly g/L
b = 0.008
state = [th[0], w[0]]

end = 24.3
points = 2430

t = [end*float(i)/(points - 1) for i in range(points)]

solutionDP = odeint(dpend, state, t)
posDP = solutionDP[:,0]
velDP = solutionDP[:,1]
```

```python
plot.figure(dpi = 100)
plot.plot(t,posDP,'b-', label = 'modeled theta')
plot.plot(t,th,'g', label = 'theta')
plot.xlabel("Time (s)")
plot.ylabel("Angle")
plot.suptitle("θ(t) and model", fontsize = 15)
plot.title("m = 0.1, k = 4.3, b = 0.008, and initial state = [th[0], w[0]]",
           fontsize = 10)
plot.legend()
plot.show()
```
