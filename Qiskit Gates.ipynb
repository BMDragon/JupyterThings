```python
from qiskit import *
```

```python
from qiskit.tools.visualization import plot_bloch_multivector
```

```python
circuit = QuantumCircuit(1, 1)
circuit.x(0)
simulator = Aer.get_backend('statevector_simulator')
result = execute(circuit, backend = simulator).result()
statevector = result.get_statevector()
print(statevector)
%matplotlib inline
circuit.draw(output='mpl')
```

```python
plot_bloch_multivector(statevector)
```

```python
circuit.measure([0], [0])
backend = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = backend, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```

```python
circuit = QuantumCircuit(1, 1)
circuit.x(0)
simulator = Aer.get_backend('unitary_simulator')
result = execute(circuit, backend = simulator).result()
unitary = result.get_unitary()
print(unitary)
circuit.draw(output='mpl')
```

```python
plot_bloch_multivector(unitary)
```

```python
circuit2 = QuantumCircuit(1, 1)
circuit2.h(0)
circuit2.x(0)
circuit2.h(0)
simulator2 = Aer.get_backend('statevector_simulator')
result = execute(circuit2, backend = simulator2).result()
statevector2 = result.get_statevector()
print(statevector2)
circuit2.draw(output='mpl')
```

```python
plot_bloch_multivector(statevector2)
```
