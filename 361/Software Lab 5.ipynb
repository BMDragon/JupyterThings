$$m\ddot{x}_1 = -kx_1 + k_2(x_2 - x_1) = -(k + k_2)x_1 + k_2x_2$$
$$m\ddot{x}_2 = -k_2(x_2 - x_1) - kx_2 = k_2x_1 - (k_2 + k)x_2$$
$$\begin{bmatrix}
m & 0\\
0 & m
\end{bmatrix}
\begin{bmatrix}
\ddot{x}_1\\
\ddot{x}_2
\end{bmatrix}
= -\begin{bmatrix}
k+k_2 & -k_2\\
-k_2 & k+k_2
\end{bmatrix}
\begin{bmatrix}
x_1\\
x_2
\end{bmatrix}$$

Case m1=m2=1, k=k2=1:
$$\begin{bmatrix}
1 & 0\\
0 & 1
\end{bmatrix}
\begin{bmatrix}
\ddot{x}_1\\
\ddot{x}_2
\end{bmatrix}
= -\begin{bmatrix}
2 & -1\\
-1 & 2
\end{bmatrix}
\begin{bmatrix}
x_1\\
x_2
\end{bmatrix}$$

Solved in question 11.9 in Homework 10:
$$x_1(t) = Acos(t-\delta) + Bcos(\sqrt{3}t-\phi)$$
$$x_2(t) = Acos(t-\delta) - Bcos(\sqrt{3}t-\phi)$$

For the symmetric mode, A = 0, B is the amplitude, and phi is the arbitraty phase (can be 0). Expected frequency is $\sqrt{3}/(2\pi)$ Hz.
For the asymmetric mode, A is the amplitude, B = 0, and delta is the arbitrary phase (can be 0). Expected frequency is $1/(2\pi)$ Hz.

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def coupled(x, t):
    x1, x2, v1, v2 = x
    dvdt = [v1, v2, -2*x1+x2, x1+-2*x2]
    return dvdt
```

```python
xsym0 = [1, -1, 0, 0]
xasym0 = [1, 1, 0, 0]
t = np.linspace(0, 10, 101)
```

```python
solSym = odeint(coupled, xsym0, t)
solAsym = odeint(coupled, xasym0, t)
```

```python
plt.figure(dpi=300)
plt.plot(t, solSym[:, 0], 'b', label=r'x_1(t)')
plt.plot(t, solSym[:, 1], 'r', label=r'x_2(t)')
plt.legend(loc='best')
plt.title('Symmetric Mode')
plt.ylabel('Displacement (m)')
plt.xlabel('time (s)')
plt.show()
```

```python
plt.figure(dpi=300)
plt.plot(t, solAsym[:, 1], 'r', label=r'x_2(t)')
plt.legend(loc='best')
plt.title('Asymmetric Mode')
plt.ylabel('Displacement (m)')
plt.xlabel('time (s)')
plt.show()
```

```python
plt.figure(dpi=300)
plt.plot(t, solAsym[:, 0], 'b', label=r'x_1(t)')
plt.legend(loc='best')
plt.title('Asymmetric Mode')
plt.ylabel('Displacement (m)')
plt.xlabel('time (s)')
plt.show()
```
