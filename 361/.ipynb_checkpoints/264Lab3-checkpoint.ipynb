```python
# importing necessary packages
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# loading data
datum = np.loadtxt('BrandonMuon', dtype=float)
mu = 1/np.mean(datum)
```

```python
# defining the histogram model
def modelHist(t, binsize):
    return len(datum)*mu*np.exp(-mu*t)*binsize
```

```python
# Making semilogy plot
plt.figure(200, dpi=300)
plt.semilogy(datum, 'b,', alpha=.5)
plt.title('Semilogy plot of pulse data')
plt.xlabel('Event number')
plt.ylabel('Elapsed time (s)')
plt.xticks(rotation=45)
plt.show(200)
```

```python
# getting initial histogram data
nBins = 400
plt.figure(0)
centers = plt.hist(datum, bins=nBins)[1]
centers = np.array([(centers[i+1]+centers[i])/2 for i in range(len(centers)-1)])
counts = plt.hist(datum, bins=nBins)[0]
plt.close(0)
Hoft = modelHist(centers, centers[1] - centers[0])
Hplus = Hoft + np.sqrt(Hoft)
Hminus = Hoft - np.sqrt(Hoft)
```

```python
# making initial histogram plot
plt.figure(1, dpi=1000)
plt.scatter(centers, counts, marker='.', alpha=.5)
plt.plot(centers, Hminus, color='orange', linewidth=.5)
plt.plot(centers, Hplus, color='orange', linewidth=.5)
plt.plot(centers, Hoft, 'r', linewidth=.5)
plt.show(1) 
```

```python
# getting log-log histogram data
plt.figure(2)
centers2 = plt.hist(np.log(datum), bins=nBins)[1]
counts2 = plt.hist(np.log(datum), bins=nBins)[0]
centers2 = np.array([(centers2[i+1]+centers2[i])/2 for i in range(len(centers2)-1)])
plt.close(2)
leftEdges = np.exp(centers2 - (centers2[1] - centers2[0])/2)
rightEdges = np.exp(centers2 + (centers2[1] - centers2[0])/2)
cents = np.array([(leftEdges[i] + rightEdges[i])/2 for i in range(len(leftEdges))])
Hoft2 = np.array([modelHist(cents[i], rightEdges[i] - leftEdges[i]) for i in range(len(cents))])
Hplus2 = Hoft2 + np.sqrt(Hoft2)
Hminus2 = Hoft2 - np.sqrt(Hoft2)
```

```python
# making log-log histogram plot
plt.figure(3, dpi=300)
plt.scatter(centers2, counts2, marker='.', alpha=.5, label='Event data')
plt.plot(np.log(cents), Hminus2, color='orange', linewidth=.5)
plt.plot(np.log(cents), Hplus2, color='orange', linewidth=.5, label=r'Model $\pm 1\sigma$')
plt.plot(np.log(cents), Hoft2, 'r', linewidth=.5, label='Model')
plt.legend()
plt.yscale('log')
plt.title('log-log plot of the data compared to the model')
plt.xlabel('Elapsed time (log(s))')
plt.ylabel('counts')
plt.show(3)
```

```python
# zooming in on log-log histogram plot
plt.figure(4, dpi=300)
plt.scatter(centers2, counts2, marker='.', alpha=.5, label='Event data')
plt.plot(np.log(cents), Hminus2, color='orange', linewidth=.5)
plt.plot(np.log(cents), Hplus2, color='orange', linewidth=.5, label=r'Model $\pm 1\sigma$')
plt.plot(np.log(cents), Hoft2, 'r', linewidth=.5, label='Model')
plt.legend()
plt.xlim(-19, -10)
plt.ylim(.01, 1000)
plt.yscale('log')
plt.title('log-log plot of the data compared to the model, Zoomed into deviation from model')
plt.xlabel('Elapsed time (log(s))')
plt.ylabel('counts')
plt.show(4)
```

```python
# filtering for only events within threshold time
newset = np.array([datum[i] for i in range(len(datum)) if np.log(datum[i]) < -11.5])
plt.figure(5)
centers3 = plt.hist(newset, bins=35)[1]
centers3 = np.array([(centers3[i+1]+centers3[i])/2 for i in range(len(centers3)-1)])
counts3 = plt.hist(newset, bins=35)[0]
plt.close(5)
Hoft3 = modelHist(centers3, centers3[1] - centers3[0])
Hplus3 = Hoft3 + np.sqrt(Hoft3)
Hminus3 = Hoft3 - np.sqrt(Hoft3)
```

```python
# plotting histogram within threshold time
plt.figure(6, dpi=300)
plt.plot(centers3, counts3, linewidth=.5, label='Event data')
plt.plot(centers3, Hminus3, color='orange', linewidth=.5)
plt.plot(centers3, Hplus3, color='orange', linewidth=.5, label=r'Model $\pm 1\sigma$')
plt.plot(centers3, Hoft3, 'r', linewidth=.5, label='Model')
plt.legend()
plt.title('Histogram of events less than the threshold time')
plt.ylabel('counts')
plt.xlabel('Elapsed time (s)')
plt.xticks(rotation=45)
plt.show(6)
```

```python
# getting statistical significance of event data
sig = (counts3[0]-Hoft3[0])/(Hplus3[0]-Hoft3[0])
print('Statistical significance: ' + str(sig) + ' sigma')
```

```python
# finding a fit function and plotting the result for the difference
diffSet = counts3 - Hoft3
fitt, cov = curve_fit(lambda t,a,b: a*b*np.exp(-b*t), centers3, diffSet,  p0=(0, 0.1))
plt.figure(7, dpi=300)
plt.scatter(centers3, diffSet, marker='.', alpha=.5, label='Difference')
plt.plot(centers3, fitt[0]*fitt[1]*np.exp(-fitt[1]*centers3), 'r', 
         linewidth=.5, label='Fitted function')
plt.xlim(-0.0000001, .0000105)
plt.legend()
plt.title('Difference between measured counts and expected counts')
plt.xlabel('Elapsed time (s)')
plt.ylabel('Difference between counts')
plt.show(7)
```

```python
# getting parameter values and uncertainties
numEvents = fitt[0]/(centers3[1] - centers3[0])
print('Number of correlated events: ' + str(numEvents) + r' $\pm$ ' + 
      str(cov[0][0]/(centers3[1] - centers3[0])))
tau = 1/fitt[1]
print('Decay time: ' + str(tau) + r' $\pm$ ' + str(1/cov[1][1]) + ' s')
```
