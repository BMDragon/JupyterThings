```python
w = np.array([])
bound = int(a*(0.5+(1/2)**0.5))
for i in range(a,bound,1):
    d = -((a**2)/2 - (i - a/2)**2)**0.5 + a/2
    if(d == int(d)):
        w = np.append(w, i)
print(w)
f = -((a**2)/2 - (w - a/2)**2)**0.5 + a/2
```

```python
figure(figsize=(8,8), dpi = 100)
plot.plot(x, y, 'b-')
plot.plot(x, z, 'b-')
plot.plot(w,f,'g.')
plot.xlim(0,bound)
plot.ylim(0,bound)
plot.show()
```
