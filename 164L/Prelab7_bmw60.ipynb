Hint: You can create an array of distance increments by taking the difference in time between each adjacent time point. This is done by array slicing: 

`distances[1:] - distances[:-1].`

Keep in mind the result will have one less element than the original array! (This may actually be exactly what you are looking for...) 



```python
import numpy as np
```

```python
def work(forces, distances):
    dx = distances[1:] - distances[:-1]
    return np.multiply(forces[0:len(dx)],dx).sum()
```

## Exercise 2.  Noise filtering

The following code cell generates an array of simulated points in time and a simulated measurement that includes noise.  

In this exercise, you will read about noise filtering.  Then you'll be modifying a code-cell containing a noise filter, and trying to obtain the best noise filtering possible.

**Please run the following cell before reading on.**  The arrays it creates are:

`tsim` --> array of times

`msim` --> array of simulated force samples

```python
import matplotlib.pyplot as plot

tstart = 0
tend = 10
npoints = 300
#Create the points in time:
tsim = np.linspace( tstart, tend, npoints)

#Calculate the corresponding "sampling rate" in points per second. To be used later.
fsample = npoints/(tend-tstart)

#Create simulated noisy force measurement.
msim = 4*np.sin(1.4*tsim+0.143) + 1.3  #Create a signal
msim += msim +np.random.randn(msim.size) #Adding pseudorandom noise.

#Plot it
plot.figure(dpi=200)
plot.plot(tsim,msim,'+')
plot.show()
```

## The noisy data above is meant to be representative of a scientific measurement in practice.  

### What can we do about the noise?

* 1.  We can create a best-fit function if all we want to do is determine a set of theoretical parameters. **This is usually the best treatment.**
* 2.  But what about the situation where we have no theoretical best-fit function to use, perhaps if the signal was of unknown origins but the noise was very well characterized?  We might then consider using a **low-pass filter**  on the data to eliminate high frequency noise (and "pass" any signal below some frequency). 

This would be particularly relevant if the signal was to be used as input to some further algorithm, such as a feedback control loop for an experiment, machine learning, or astronomical image processing, for instance.  

*Notice that the oscillating signal shown above varies over a timescale of seconds, but the noise varies over a much shorter time scale.* This is frequently the case with noise in measurement systems (but not always!).   We can use this to our advantage by apply a smoothing method that eliminates noise above some desired frequency threshold, but keeps signals below that threshold.

## Noise filtering

**Rule number one for noise filtering:** Don't use a simple sliding-average.  Although taking the average of successive groups of points within your data might seem like a nice, simple idea, it will distort any and all signals within your data as well.  The only situation where you should use a simple average is when your data represents a *constant, unchanging quantity.*

**Rule number two: filter when your noise falls in a different frequency regime from your signal**.   If your noise is very low frequency, but your signal is fast-changing (high frequency) then this works.  If your signal is in a lower frequency regime (slowly changing compared to your noise) then this also works -- and that is a very common situation.

One of the most common methods of noise filtering used in science is the "**Butterworth filter**".  This filter offers a well defined "pass band" -- a range of frequencies over which all signals will be unchanged;  then frequencies above the "pass band" are successively reduced, filtered away. The "cutoff frequency" is the frequency -- the rate of change -- at which the pass band ends, above which all signals are reduced.

Achieving an ideal "best" smoothing is something of a tricky science with a great deal of mathematics behind it.  In practice you must have some criterion to decide what your acceptable tradeoff might be:  more filtering runs the risk of removing important signal information as well, but reduces noise more completely; less smoothing leaves your signal relatively undistorted but likewise leaves it embedded in noise.  

### For a first-pass at noise filtering in this course, we will try to develop in intuition for what over-smoothing looks like and what smoothing can do to your data in a qualitative sense.

```python
### Noise filtering function:  Create a Butterworth filter

fc = 5  # Cut-off frequency of the filter in Hz.
# Must be LESS THAN HALF OF THE SAMPLING FREQUENCY!

cutoff = fc / (fsample / 2) # Normalize the frequency for our sampling rate.  A necessary step due to how the filter is implemented.

#Cutoff now represents our required cutoff frequency.

#Create a Butterworth filter.  
from scipy import signal  #Use the signal processing SciPy module.

#Create a Butterworth filter object we can use.
#  Format:   scipy.signal.butter( order , cutoff , btype='low' )
#
#  Order:  the 'order' of the filter.  Higher gives a steeper cutoff, but isn't always better.
# Cutoff:  the cutoff frequency;  frequencies above this will be successively reduced.
order = 4

b, a = signal.butter( order , cutoff, 'low')

# b, a  are the filter parameters SciPy can use to implement the noise filtering we need.
# These are technically "parameters of an Infinite Impulse Response filter" and knowing what that means is an important 
# branch of science and engineering in itself!  The science is called, "Signals and Systems."

#Now do the filtering on our simulated force measurement!
output = signal.filtfilt(b, a, msim)

#Plot the result
plot.figure(dpi=200)
plot.plot(tsim, msim, 'r.',label="noisy")
plot.plot(tsim, output, 'b-', label='filtered')
plot.xlabel("Time (s)")
plot.ylabel("Force")
plot.legend()
plot.show()
```
