```python
import numpy as np
import matplotlib.pyplot as plot
from scipy.integrate import odeint
```

```python
def driven(stateVec, t):
    x,v = stateVec
    f = [v, F_0 * np.sin(freq * t) - (b*v/m) - (k*x/m)]
    return f
```

```python
#Driven
m = 3.14
k = 1.618
b = 0.01
F_0 = 0.1
freq = 1
state = [0.0, 0.0]

end = 100.0
points = 96000

t = [end*float(i)/(points - 1) for i in range(points)]

solutionD = odeint(driven, state, t)
posD = solutionD[:,0]
velD = solutionD[:,1]
```

```python
plot.figure(dpi = 100)
plot.plot(t,posD,'b-', label = 'driven position (b = 0.01)')
plot.plot(t,velD,'b:', label = 'driven velocity')
plot.xlabel("Time (s)")
plot.ylabel("Position (m) and Velocity (m/s)")
plot.suptitle("x(t) and v(t) for damped driven SHO", fontsize = 15)
plot.title("m = 3.14, k = 1.618, freq = 1, F_0 = 0.1 and initial state = [0.0, 0.0]",
           fontsize = 10)
plot.legend()
plot.show()
```
