Computer Models Using Equations of Motion

Introduction
-------------
By now you should be familiar with the concept of an equation of motion for a system.  This is a *differential* equation (usually a second order one, since acceleration is the second derivative of position) that is typically derived by examining the net force (sum of forces) acting on a system.  (In more advanced treaments an equation of motion may instead be derived using Lagrangian or Hamiltonian methods.)

$$ \vec{F}_{\text{net}} = \Sigma_i \vec{F}_i = m\vec{a} = m\frac{\text{d}^2 \vec{x}}{\text{d}t^2}
$$

The typical process for solving physics problems is, often enough, to first somehow find the equation of motion for a system, then to solve the equation for position, velocity, momentum or other required things.  This typically entails integration. 

## Simulating physics:  Numerical integration of equations of motion

**NOTE:  Here we are talking about integrating a *system of differential equations* to find a close approximation of its solution; you should consider this as distinct from the "numerical integration" we talked about last week, which simply involved how to take a Riemann sum for a data set!**  

You can use a computer to *numerically integrate* an equation of motion, effectively solving numerically for $x(t)$ and $v(t)$, for example.  This is in effect a **computer model** of the system in question, a simulation.  

### Euler integration: the first and most simple method

Euler integration is still taught because its simplicity makes clear how the process works.  It is not terribly accurate compared to newer methods, but for simple systems it actually works quite well! The process goes as follows:

1. Your system must start at some initial position and velocity.  This is the "initial state" of the system, and you need these values to begin with.  So write down $(x_0, v_0)$.

2. Write down the acceleration of the system, $a$, from the equation of motion.

3. Specify a time increment $\Delta t$ and a start and end time, $t_0$ and $t_{\text{end}} to integrate. 

4. *The integration process goes like this*:

    * Update t:  $t_i = t_{i-1} + \Delta t$
    * Update v:  $v_i = v_{i-1} + a*\Delta t$
    * Update x:  $x_i = x_{i-1} + v_{i-1}*\Delta t$
    * Store $(t_i, x_i, v_i)$
    * Now $t_i$ becomes the new $t_{i-1}$.  Repeat until $t_i \ge t_{\text{end}}$.

In this process, it is worth noting that $a$ need not be a constant quantity; it may well be a function e.g. of $t_{i-1}$ or $x_{i-1}$ for example.

#### The main theme here is that infinitesimals have been approximated by finite differences:  $ \text{d}x \rightarrow \Delta x$, $\text{d}t \rightarrow \Delta t$, etc.  

**This gives you the basic idea of how to integrate a differential equation.  However more accurate methods exist; for this reason we will tend to first try the built-in routine in the SciPy package.**



Numerical Integration using `odeint()`
-----------------------------------

The SciPy package `scipy.integrate` includes the `odeint()` command which contains a sophisticated numerical integration routine meant specifically for integrating systems of differential equations.  Other computer languages, such as Matlab, include similar commands that all function in about the same way.  Thus you should consider that the following tutorial is fairly general across most computer languages used for computer modeling.

The `odeint()` command and similar commands in other languages rely on the fact that differential equations can be broken up into systems of *first order* differential equations and then treated using *vector algebra*.  This is much more straightforward than it sounds, and the best way to learn how it's done is by looking at an example.  

**This Notebook contains an example of how to create a computer model for a simple harmonic oscillator, a mass on a spring.**  Before we look at the computer code, we first have to come up with our equation of motion and see how to represent it using vector algebra.  Please read on.

# Simulation example/ tutorial:  A Mass-Spring System

### Equation of motion

For a mass on a horizontal spring in absence of friction or drag, the sum of forces acting on the mass *m* will look like this:

\begin{equation*}
    F = -kx(t) = ma(t) 
\end{equation*}

Thus, in standard form:

\begin{equation*} 
    \frac{\text{d}^2 x}{\text{d}t^2} + \frac{k}{m}x(t) = 0 
\end{equation*}

### Transform into a system of *first order* differential equations

We can transform this formally into a system of first order differential equations by defining "stand-in" functions for each derivative.  In this case we simply define a function *v(t)* to be the first derivative of *x(t)*:

\begin{align} \text{define}\; \frac{\text{d}x}{\text{d}t}&= v(t)\end{align}

Thus we have a system of two first order equations from one second order equation:

\begin{eqnarray}
  \frac{\text{d}x}{\text{d}t}&=&v(t)\\\\
\frac{\text{d}v}{\text{d}t}&=&\frac{-k}{m}x(t) 
\end{eqnarray}

### Vector algebraic representation

What this means is that we can define a **state vector** for this system, which is simply the vector of the two functions we used to make our system of first order differential equations:  *x(t)* and *v(t)*. 

\begin{equation*}
    \vec{q}(t) = \big[ x(t),\: v(t) \big]
\end{equation*}

We refer to this as a state vector because at any given instant it contains all of the information necessary to totally and *uniquely* specify the state of the system being studied (together with the values of the constant parameters, of course)!  Note that if the system were 2-dimensional, then the state vector would include x, y, and the x and y components of velocity: four different functions.

Then the derivative of our state vector is:

\begin{equation*}
\frac{\text{d}\vec{q}}{\text{d}t} = \big[ \frac{\text{d}x}{\text{d}t},\:\frac{\text{d}v}{\text{d}t} \big] = \big[ v(t), \frac{-k}{m}x(t) \big]
\tag{1}
\end{equation*}

### Equation (1) above is what we need to utilize `odeint()`!

The following is not strictly necessary to know, but if you want to take it farther, for the mass-spring system, the matrix form of our equation of motion has become this:

\begin{align}
\frac{\text{d}\vec{q}(t)}{\text{d}t}= \begin{bmatrix} 0& 1\\\\  -\frac{k}{m}& 0 \end{bmatrix} \vec{q}(t) 
\end{align}

Now all we require is to define a *Python function* that takes as input a state vector and returns as output the derivative, which you obtain simply by multiplyiing that state vector by the matrix!  Once you specify *k* and *m* as numerical values, such a python function simply takes in a set of numbers and transforms them into a new set of numbers, returning that.  Below, you'll see how this is done.

### Initial conditions

As explained previously, numerical integration must *begin* from some initial conditions in order to produce a meaningful result.  All computers models of physical systems must explicitly specify the initial conditions for the simulation!  This is done by **specifying initial values for each function in the state vector**.  Thus your system begins from some initial position $x_0$ at some initial velocity $v_0$, specified as an initial state vector $\vec{q}_0$.

## A word about accuracy and instability

The degree to which the computer model matches reality depends upon several factors:

1. The degree to which your equation of motion actually matches the real physical system's behavior. 

2. The precision and accuracy with which you know the physical constants and parameters (such as mass, spring constants, &c.) used by the equation of motion.

3. The precision and accuracy with which your computer model's *initial conditions* match those that physically occur.

4. The time *step size* with which you numerically integrate.  Smaller time slices allow greater accuracy at a cost in processing time and memory required to complete the simulation (due to greater RAM and CPU utilization).

5. The *numerical precision* with which you represent numbers.  Calculations done by a computer may be subject to truncation or round-off errors due to a finite number of decimal places being used to represent quantities.  This is particularly relevant when you try to study nonlinear systems, which may depend quite sensitively on information contained at all levels of significance after the decimal place.

6. **When it fails:** When the time-step for integration is too large, or when the system parameters change nonlinearly and very fast, the computation can experience a **numerical instability**, usually resulting in your position or velocity rapidly escaping to infinity or -infinity.  This is a sign you must be more careful in your simulation, reduce the time step and increase settings for precision.

```python
#IMPORT STANDARD MODULES
import numpy as np
import matplotlib.pyplot as plot
import matplotlib as mpl
mpl.rc('figure',dpi=180)  #For bigger plots
```

The following import statement loads the `odeint()` function we'll require to numerically integrate our system:

```python
from scipy.integrate import odeint
```

Next, specify the values of the physical parameters and constants:

```python
#Our system has only mass and spring constant
m = 1.0
k = 3.3
```

Initial conditions can be grouped in an array that represents the initial state-vector of the system:

```python
#Initial conditions [x0,v0]
#Let's set the mass initially centered at equilibrium position but moving at 0.5 m/s . 
statevector0 = [0.0 ,0.5]
```

The following function accepts a state vector and returns the derivative of it!  That is all the `odeint()` function requires to perform numerical integration.  **Note that we have to pass in the time array, t** as a formality.  This is a result of the way that `odeint()` is designed, but has no other relevance for us.

```python
def eqn_of_motion(statevector, t):
    #First unpack the values stored in the vector.
    x,v = statevector
    
    #Create f = [dx/dt, dv/dt]
    #This is in fact all we need to do: make the return vector 
    #component by component.  We do not need to use matrix multiplication
    #or anything fancy at all, because we can do that on paper ahead of time
    #and just type in the result in component form:
    f = [v, - (k/m)*x]
    
    return f
```

**Time range and number of time steps** are defined below:

```python
t_end = 10.0  #Seconds; we assume time starts at t = 0 without loss of generality.
numpoints = 5000 #Number of time steps to integrate between t = 0 and t = t_end
#We will use this to calculate the time step between each consecutive point based on the the total time spend (t_end).
#Note that making numpoints GREATER will increase the accuracy of our numerical solution, but at some point may bog
#down the computer!
```

The following command creates an array with as many time values as "numpoints" specifies, between 0 and t_end.  If you do not completely see how this works, don't worry about it yet.  You may re-use this line of code in most models you are likely to produce. In effect this statement is implicitly creating a list of numbers and assigning that list to "t".

```python
t = [t_end*float(i) / (numpoints - 1) for i in range(numpoints)]
```

The following statement *actually integrates the system* and creates lists of numbers for x(t) and v(t), the numerical solutions.  These are stored in a two-column list "sol1".  We call the `odeint()` command to integrate our system over the time range we have defined.

```python
sol1 = odeint(eqn_of_motion, statevector0, t)
positions = sol1[:,0]
velocities = sol1[:,1]
```

If you look at the structure of the solution array, `sol1`, you'll see two columns.  The first column corresponds to x(t) and you can use *array slicing* to refer to it, sol1[:,0] referes to the first column, for example.  In the statement above, we have stored each column into a new Numpy Array with an easy name!  Below, let's plot *both columns*, the position and the velocity.

The columns created by `odeint()` will correspond exactly to the number of functions that form the state vector of the system,  thus one numerical solution is created per function required to be solved!  In this case we needed solutions for velocity v(t) and also position x(t).  If our system contained several variables (for example coupled oscillators with multiple moving masses), the solution *and* the state vector would have more columns and more functions to be solved-for respectively!

```python
plot.plot(t, positions,"g")  #g  means green
plot.plot(t, velocities,"b:")  #b:  means blue and dotted
plot.xlabel("Time (s)")
plot.ylabel("Position (green) / Velocity (blue)")
plot.title("x(t) and v(t) for model SHO, mass = 1kg, k = 3.3 N/m")
# YOUR OWN MODEL SHOULD USE LABELED AND TITLED PLOTS!
```

```python
def damped(stateVec, t):
    x,v = stateVec
    f = [v, -(b*v/m) - (k*x/m)]
    return f
```

```python
#Underdamped
m = 2.718
k = 1.618
b = 0.314*(4*m*k)**0.5
state = [0.0, .5]

end = 25.0
points = 5000

t = [end*float(i)/(points - 1) for i in range(points)]

solutionU = odeint(damped, state, t)
posU = solutionU[:,0]
velU = solutionU[:,1]
```

```python
#Overdamped
b = 4.2*(4*m*k)**0.5

solutionO = odeint(damped, state, t)
posO = solutionO[:,0]
velO = solutionO[:,1]
```

```python
#Critically damped
b = (4*m*k)**0.5

solutionC = odeint(damped, state, t)
posC = solutionC[:,0]
velC = solutionC[:,1]
```

```python
plot.figure(dpi = 100)
plot.plot(t,posU,'b-', label = 'underdamped position (b = 0.314(4mk)^0.5)')
plot.plot(t,velU,'b:', label = 'underdamped velocity')
plot.plot(t,posO,'g-', label = 'overdamped position (b = 4.2(4mk)^0.5)')
plot.plot(t,velO,'g:', label = 'overdamped velocity')
plot.plot(t,posC,'r-', label = 'critically damped position (b = (4mk)^0.5)')
plot.plot(t,velC,'r:', label = 'critically damped velocity')
plot.xlabel("Time (s)")
plot.ylabel("Position (m) and Velocity (m/s)")
plot.suptitle("x(t) and v(t) for damped SHO", fontsize = 15)
plot.title("m = 2.718, k = 1.618, and initial state = [0.0, 0.5]")
plot.legend()
plot.show()
```
